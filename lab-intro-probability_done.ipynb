Challenge 1 — Ironhack Airlines

```python
from scipy.stats import binom

n = 460
p = 0.97
binom_dist = binom(n, p)

prob_all_seated = binom_dist.cdf(450)
print(prob_all_seated)
```

Challenge 2 — Ironhack Call Center

```python
from scipy.stats import geom

p = 0.3
geom_dist = geom(p)

prob_at_least_three = 1 - geom_dist.cdf(2)
print(prob_at_least_three)
```

Challenge 3 — Ironhack Website

```python
from scipy.stats import poisson

mu = 500
poisson_dist = poisson(mu)

prob_overwhelmed_hour = 1 - poisson_dist.cdf(550)
print(prob_overwhelmed_hour)

prob_overwhelmed_day = 1 - (poisson_dist.cdf(550))**24
print(prob_overwhelmed_day)
```

Challenge 4 — Ironhack Helpdesk

```python
from scipy.stats import expon

lambda_value = 1/10
exp_dist = expon(scale=1/lambda_value)

prob_within_5 = exp_dist.cdf(5)
print(prob_within_5)

prob_break = 1 - exp_dist.cdf(15)
print(prob_break)
```

Challenge 5 — Birds Weights

```python
from scipy.stats import norm

mean = 150
std = 10
norm_dist = norm(loc=mean, scale=std)

prob_140_160 = norm_dist.cdf(160) - norm_dist.cdf(140)
print(prob_140_160)
```

Challenge 6 — Component Lifetime

```python
from scipy.stats import expon

mean_lifetime = 50
exp_dist = expon(scale=mean_lifetime)

prob_fail_30h = exp_dist.cdf(30)
print(prob_fail_30h)
```
